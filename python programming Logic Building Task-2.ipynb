Problem 1: Unique Words in a Sentence

```python
sentence=input("Enter sentence: ")
words=sentence.split()
unique_words=set(words)
print("Unique words count:",len(unique_words))
print("Unique words:",unique_words)
```

Problem 2: Highest Salary from Employee Data

```python
n=int(input("Enter number of employees: "))
employees={}
for i in range(n):
    name=input("Enter employee name: ")
    salary=int(input("Enter salary: "))
    employees[name]=salary
highest_employee=max(employees,key=employees.get)
print("Highest Salary:",highest_employee,"-",employees[highest_employee])
```

Problem 3: Find Maximum and Minimum Values

```python
n=int(input("Enter number of elements: "))
numbers=[]
for i in range(n):
    numbers.append(int(input("Enter number: ")))
max_val=numbers[0]
min_val=numbers[0]
for num in numbers:
    if num>max_val:
        max_val=num
    if num<min_val:
        min_val=num
print("List:",numbers)
print("Max:",max_val)
print("Min:",min_val)
```

Problem 4: Count Products Above a Price Threshold

```python
n=int(input("Enter number of products: "))
prices=[]
for i in range(n):
    prices.append(int(input("Enter price: ")))
count=0
for price in prices:
    if price>1000:
        count+=1
print("Products above 1000:",count)
```

Problem 5: Calculate Attendance Percentage

```python
n=int(input("Enter number of days: "))
attendance=[]
for i in range(n):
    attendance.append(input("Enter P/A: "))
present_count=0
for status in attendance:
    if status=="P":
        present_count+=1
percentage=(present_count/len(attendance))*100
print("Attendance Percentage:",percentage)
```

Problem 6: Remove Duplicate Phone Numbers

```python
n=int(input("Enter number of phone numbers: "))
phone_numbers=[]
for i in range(n):
    phone_numbers.append(int(input("Enter phone number: ")))
unique_numbers=set(phone_numbers)
print("Unique phone numbers:",unique_numbers)
```

Problem 7: Count Character Frequency

```python
text=input("Enter a string: ")
frequency={}
for ch in text:
    if ch in frequency:
        frequency[ch]+=1
    else:
        frequency[ch]=1
print(frequency)
```

Problem 8: Convert List to Tuple

```python
n=int(input("Enter number of elements: "))
lst=[]
for i in range(n):
    lst.append(int(input("Enter element: ")))
t=tuple(lst)
print("Tuple:",t)
```

Problem 9: Check if a Key Exists in a Dictionary

```python
n=int(input("Enter number of employees: "))
employees={}
for i in range(n):
    name=input("Enter employee name: ")
    salary=int(input("Enter salary: "))
    employees[name]=salary
key=input("Enter employee name to search: ")
if key in employees:
    print("Employee exists")
else:
    print("Employee does not exist")
```

Problem 10: Calculate Average Marks

```python
n=int(input("Enter number of subjects: "))
marks=[]
for i in range(n):
    marks.append(float(input("Enter mark: ")))
total=0
for m in marks:
    total+=m
average=total/len(marks)
print("Average Marks:",average)
```
